```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## Load image

```python
filename = 'imgs/chicken.jpg'
```

### Deep Learning model + Pre-Processing

```python
mobile = tf.keras.applications.mobilenet.MobileNet() # model 1
mobile2 = tf.keras.applications.mobilenet_v2.MobileNetV2() # model 2
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename,target_size=(224,224))
plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img,axis=0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
resized_img.shape #RBG + 3dim
```

```python
final_image.shape #RGB + 3dim + 1dim
```

```python
predictions = mobile.predict(final_image)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
for inner_list in results:
    for item in inner_list:
        name = item[1]
        score = item[2]
        print(name,score)
```

```python
predictions2 = mobile2.predict(final_image)
results2 = imagenet_utils.decode_predictions(predictions2)
print(results2)
```

```python
for inner_list in results2:
    for item in inner_list:
        name = item[1]
        score = item[2]
        print(name,score)
```
